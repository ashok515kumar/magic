```python
def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle 

def process_video(over_time, extract_landmarks, list_images):
    flag = False
    count = 0
    for i in range(len(list_image)):
        landmarks = list_results[i].pose_landmarks.landmark
        shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
        elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
        wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

        angle = calculate_angle(shoulder, elbow, wrist)
        
    
        if (angle > 20):
            flag = True
                
        if (flag == True and angle < 12.5):
            count += 1
            flag = False
            
        # Visualize angle
        cv2.putText(list_image[i], "Percentage of Wave", 
                               tuple(np.multiply(elbow, [250, 480]).astype(int)), 
                               cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                    )

        cv2.putText(list_image[i], str(count/org_count * 100), 
                tuple(np.multiply(elbow, [640, 480]).astype(int)), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                )

        cv2.putText(list_image[i], "Wave Count :", 
                               tuple(np.multiply(elbow, [400, 580]).astype(int)), 
                               cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                    )

        cv2.putText(list_image[i], str(count), 
                tuple(np.multiply(elbow, [640, 580]).astype(int)), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                )

        mp_drawing.draw_landmarks(list_image[i], list_results[i].pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                     mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2),
                                     mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2))
        print(list_results[i].pose_landmarks)
        cv2.imshow('Mediapipe Feed', list_image[i])
        
        
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
            
    cv2.destroyAllWindows()
    

cap = cv2.VideoCapture('B.mp4')
landmarks = 0
angle_vals = []
flag = False
list_landmarks = []
list_image = []
list_image_bgr = []
list_results = []
org_count = 0
## Setup mediapipe instance
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while(cap.isOpened()):
        ret, frame = cap.read()
        
        if ret == False:
            break
        # Convert into Gray
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        list_image_bgr.append(image)
        results = pose.process(image)
        
        image.flags.writeable = True
        image = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
        list_image.append(image)
        list_results.append(results)
        # Extract Landmarks
        try:
            landmarks = results.pose_landmarks.landmark
            list_landmarks.append(landmarks)
            
            # Get coordinates
            shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
            elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
            wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

            angle = calculate_angle(shoulder, elbow, wrist)
            angle_vals.append(angle)
            if (angle > 20):
                flag = True
                
            if (flag == True and angle < 12.5):
                org_count += 1
                flag = False
                
            
            
        except:
            pass
            
    cap.release()
    cv2.destroyAllWindows()
print(org_count)
process_video(org_count, list_results, list_image)
```
